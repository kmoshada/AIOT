```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
try:
    df = pd.read_csv('/home/km/Developments/Python/AIOT/Clean Datasets/nafld_cleaned.csv')

    # --- Create a figure with multiple subplots ---
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle('Analysis of NAFLD Dataset Features', fontsize=16)

    # 1. Age distribution by status
    sns.histplot(data=df, x='age', hue='status', multiple='stack', ax=axes[0])
    axes[0].set_title('Age Distribution by Status')

    # 2. BMI distribution by status
    sns.histplot(data=df, x='bmi', hue='status', multiple='stack', ax=axes[1])
    axes[1].set_title('BMI Distribution by Status')

    # 3. Gender distribution by status
    sns.countplot(data=df, x='male', hue='status', ax=axes[2])
    axes[2].set_title('Gender Distribution by Status')
    axes[2].set_xticklabels(['Female', 'Male'])

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

    # --- Create a Pair Plot for an overview of relationships ---
    print("\nGenerating pair plot...")
    # A pair plot shows pairwise relationships between variables.
    # The diagonal shows the distribution of each variable.
    pair_plot = sns.pairplot(df, hue='status', palette='viridis')
    pair_plot.fig.suptitle('Pairwise Relationships by Status', y=1.02)
    plt.show()


except FileNotFoundError:
    print("Error: The file '/home/km/Developments/Python/AIOT/Clean Datasets/nafld_cleaned.csv' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")

```
